# 1.3 `Train_Test_Split: `

- In sk-learn we have `train_test_split()`.
- But for tensorflow we have no inbuid function but we use `.take()`. 

<br>

```python

# training data set we have: 130*0.8 = 104 images for traning : 
train_size = int(total_img_length * 0.8)

train_ds = image_ds.take(train_size)
test_ds = image_ds.skip(train_size)

# print the length of train_ds and test_ds:

print("length of train_ds: {}".format(len(train_ds)))
print(f"length of test_ds {len(test_ds)}")
```

## First,  let know about string path difference in linux and windows  operating system. 


`os.path.sep` is a string attribute in Python that represents the operating system's directory separator (such as `/` on Unix-like systems, including Linux and macOS, and `\` on Windows).

It helps when you need to write cross-platform code that deals with file paths, ensuring that you use the correct directory separator for the operating system your code is running on.

### Output:
- **On Unix-like systems**: `user/documents/file.txt`
- **On Windows**: `user\documents\file.txt`

Using `os.path.sep` ensures your code handles file paths correctly across different operating systems.

<br>
<br>

```python

def getLevel(path):
    import os
    return tf.strings.split(path,os.path.sep)[-2]

for i in image_ds: 
    print(getLevel(i))
    print(getLevel(i).numpy())
    print()
    
```

```python

# we can also use map function : 

for i in train_ds.map(getLevel):
    print(i.numpy())
    
    
```

# 1.5 `Prepare Traning data:`

- In our train data we have input and output.
- For image: in `X_train` it should be image.
- For image: in `Y_train` it should be label.



<br>

```python

def process_image(file_path):
    lebel = getLevel(file_path)
    # tf.io.read_file() this function store the image:
    img = tf.io.read_file(file_path)
    # our image is jpeg we need to decode it.
    img = tf.io.decode_jpeg(img)
    # image should be different dimention we reduce the dimention of the image
    img = tf.image.resize(img,[128,128])
    return img, lebel
```

```python


for image,level in train_ds.map(process_image).take(3):
    print(f"image: {image}")
    print(f"Lebel: {level}")
    
```


<br>

# 1.6 `Scaling: in our train data set: `

<br>


```python

def scaling(image,level):
    return image/255,level
```

```python

df_train = train_ds.map(process_image).map(scaling)

for image,level in df_train.take(5):
    print(image[0][0])
    print(level)
    print()
```
